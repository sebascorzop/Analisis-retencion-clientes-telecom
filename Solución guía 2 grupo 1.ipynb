#**Análisis de Churn y Factores Relacionados**

###Calcula el porcentaje de clientes que han desertado (churn = 1).

```python
# Importamos la librería Pandas para manipular datos
import pandas as pd

# Contamos el total de clientes sin filtrar la columna "churn"
total_clientes = df["churn"].shape[0]
print(f"El total de clientes es de {total_clientes}")

# Contamos cantidad de desertores y no desertores
desertores = df[df["churn"] == True].shape[0]
no_desertores = df[df["churn"] == False].shape[0]
print(f"El total de clientes que continúan en la compañía es de: {no_desertores}\n"
      f"El total de clientes que han desertado es de: {desertores}")

# Calculamos el porcentaje de desertores dividiendo el total de desertores entre el total de clientes
porcentaje_desertores = (desertores / total_clientes) * 100
print(f"El porcentaje de clientes que han desertado es de {porcentaje_desertores:.2f}%")

# prueba groupby para corroborar hallazgos
# agrupar_por_churn = df.groupby("churn").size()
# agrupar_por_churn_desertores = df[df["churn"] == True].groupby("churn").size()
# print(agrupar_por_churn)
# print(agrupar_por_churn_desertores)
```

###Identifica si los clientes con plan internacional (international plan) tienen mayor tasa de deserción.

```python
# Importamos la librería Pandas para manipular datos
import pandas as pd

# Contamos el total de clientes sin filtrar la columna "churn"
total_clientes = df["churn"].shape[0]
print(f"El total de clientes es de {total_clientes}")

# Contamos clientes con y sin plan
con_plan = df[df["international plan"] == "yes"].shape[0]
porcentaje_con_plan = (con_plan / total_clientes) * 100
sin_plan = df[df["international plan"] == "no"].shape[0]
porcentaje_sin_plan = (sin_plan / total_clientes) * 100
print(f"La cantidad de clientes con plan internacional es de: {con_plan} y corresponde al: {porcentaje_con_plan:.2f}%")
print(f"La cantidad de clientes sin plan internacional es de: {sin_plan} y corresponde al: {porcentaje_sin_plan:.2f}%")

# Contamos cantidad de desertores con y sin plan
desertores = df[df["churn"] == True].shape[0]
desertores_sin_plan = df[(df["international plan"] == "no") & (df["churn"] == True)].shape[0]
desertores_con_plan = df[(df["international plan"] == "yes") & (df["churn"] == True)].shape[0]
porcentaje_desertores_sin_plan = (desertores_sin_plan / sin_plan) * 100
porcentaje_desertores_con_plan = (desertores_con_plan / con_plan) * 100
print(f"""El total de clientes que han desertado es de {desertores}
El total de clientes desertores con plan internacional es de: {desertores_con_plan} y equivale al {porcentaje_desertores_con_plan:.2f}%
El total de clientes desertores sin plan internacional es de: {desertores_sin_plan} y equivale al {porcentaje_desertores_sin_plan:.2f}%""")

# Condicional comparando % desertores con y sin plan internacional
if porcentaje_desertores_con_plan > porcentaje_desertores_sin_plan:
  print(f"La tasa de deserción de usuarios con plan internacional es mayor")
elif porcentaje_desertores_con_plan < porcentaje_desertores_sin_plan:
  print(f"La tasa de deserción de usuarios sin plan internacional es mayor")

# prueba groupby para corroborar hallazgos
# agrupar_por_plan = df.groupby("international plan").size()
# agrupar_por_plan_desertores = df[df["churn"] == True].groupby("international plan").size()
# print(agrupar_por_plan)
# print(agrupar_por_plan_desertores)
```

###Compara la cantidad de minutos usados en llamadas diurnas (total day minutes) entre clientes con y sin churn.

```python
import pandas as pd  # Importamos la librería Pandas para manipulación de datos

# Contamos el total de clientes sin filtrar la columna "churn"
total_clientes = df["churn"].shape[0]

# Filtramos  los clientes que han desertado y los que no
filtro_desertores = df[df["churn"] == True]
filtro_no_desertores = df[df["churn"] == False]

# Contamos la cantidad de clientes en cada grupo
desertores = filtro_desertores.shape[0]
no_desertores = filtro_no_desertores.shape[0]

# Calculamos los minutos promedio de llamadas diurnas
minutos_llamadas_diurnas_total = df['total day minutes'].sum()
minutos_llamadas_diurnas_desertores = filtro_desertores['total day minutes'].sum()
minutos_llamadas_diurnas_no_desertores = filtro_no_desertores['total day minutes'].sum()

# Mostramos los resultados
print(f"La cantidad de minutos usados en llamadas diurnas por los {total_clientes} clientes es de: {minutos_llamadas_diurnas_total:.2f} minutos")
print(f"La cantidad de minutos usados en llamadas diurnas por los {desertores} clientes desertores es de: {minutos_llamadas_diurnas_desertores:.2f} minutos")
print(f"La cantidad de minutos usados en llamadas diurnas por los {no_desertores} clientes vigentes es de: {minutos_llamadas_diurnas_no_desertores:.2f} minutos")

# prueba groupby para corroborar hallazgos
# print(df["total day minutes"].sum())
# print(df.groupby("churn")["total day minutes"].sum())
```

###Compara el número total de llamadas (total day calls) entre clientes con y sin churn.


```python
import pandas as pd  # Importamos la librería Pandas para manipulación de datos

# Contamos el total de clientes
total_clientes = df.shape[0]

# Filtramos los clientes que han desertado y los que no
filtro_desertores = df[df["churn"] == True]
filtro_no_desertores = df[df["churn"] == False]

# Contamos la cantidad de clientes en cada grupo
desertores = filtro_desertores.shape[0]
no_desertores = filtro_no_desertores.shape[0]

# Calculamos la cantidad total de llamadas diurnas
total_llamadas = df['total day calls'].sum()
total_llamadas_desertores = filtro_desertores['total day calls'].sum()
total_llamadas_no_desertores = filtro_no_desertores['total day calls'].sum()

# Calculamos el promedio de llamadas por cliente en cada grupo
promedio_llamadas_desertores = total_llamadas_desertores / desertores if desertores > 0 else 0
promedio_llamadas_no_desertores = total_llamadas_no_desertores / no_desertores if no_desertores > 0 else 0

# Mostramos los resultados
print(f"Total de llamadas realizadas por todos los clientes {total_clientes} : {total_llamadas}")
print(f"Total de llamadas realizadas por clientes {desertores} desertores: {total_llamadas_desertores}")
print(f"Total de llamadas realizadas por clientes {no_desertores } no desertores: {total_llamadas_no_desertores}")

# prueba groupby para corroborar hallazgos
print(df["total day calls"].sum())
print(df.groupby("churn")["total day calls"].sum())
```
